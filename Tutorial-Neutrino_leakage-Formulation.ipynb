## Step 1.a: General relativistic magnetohydrodynamics \[Back to [Top](#toc)\]
$$\label{grmhd_equations}$$

The standard general relativistic [magnetohydrodynamics](https://en.wikipedia.org/wiki/Magnetohydrodynamics) (GRMHD) are given by (see e.g., [Duez *et al.* (2005)](https://arxiv.org/pdf/astro-ph/0503420.pdf) for an excellent overview)

$$
\newcommand{\RR}{\mathcal{R}}
\newcommand{\QQ}{\mathcal{Q}}
\newcommand{\nui}{\nu_{i}}
\newcommand{\anui}{\bar{\nu}_{i}}
\newcommand{\nux}{\nu_{\rm x}}
\newcommand{\anux}{\bar{\nu}_{\rm x}}
\newcommand{\nue}{\nu_{\rm e}}
\newcommand{\anue}{\bar{\nu}_{\rm e}}
\newcommand{\mb}{m_{\rm b}}
\newcommand{\nb}{n_{\rm b}}
\newcommand{\ne}{n_{\rm e}}
\newcommand{\rhob}{\rho_{\rm b}}
\newcommand{\ye}{Y_{\rm e}}
\newcommand{\Fdual}{{}^{*}\!F}
\newcommand{\ee}{e^{-}}
\newcommand{\ae}{e^{+}}
\begin{align}
\nabla_{\mu}\left(\nb u^{\mu}\right) &= 0,\\
\nabla_{\mu}T^{\mu\nu} &= 0,\\
\nabla_{\nu}\Fdual^{\mu\nu} &= 0.
\end{align}
$$

These are conservation of baryon number, energy and momentum conservation, and homogeneous [Maxwell's equations](https://en.wikipedia.org/wiki/Maxwell%27s_equations), respectively. In the above, $\nabla_{\mu}$ is the covariant derivative associated with the spacetime metric $g_{\mu\nu}$, $\nb$ is the baryon number density, $u^{\mu}$ is the fluid four-velocity, and $\Fdual^{\mu\nu} = \frac{1}{4}\epsilon^{\mu\nu\rho\sigma}F_{\rho\sigma}$ is the dual of the [Faraday tensor](https://en.wikipedia.org/wiki/Electromagnetic_tensor) $F^{\mu\nu}$ ($\epsilon^{\mu\nu\rho\sigma}$ is the [Levi-Civita tensor](https://en.wikipedia.org/wiki/Levi-Civita_symbol)). $T^{\mu\nu}$ is the energy-momentum tensor of a magnetized fluid and is given by (cf. Eq. 8 of [Etienne *et al.*(2015)](https://arxiv.org/pdf/1501.07276.pdf))

$$
T^{\mu\nu}_{\rm GRMHD} = \left(\rhob h + b^{2}\right)u^{\mu}u^{\nu} + \left(P + P_{\rm mag}\right)g^{\mu\nu} - b^{\mu}b^{\nu},
$$

where $h = 1 + \epsilon + P/\rhob$ is the specific enthalpy, $\epsilon$ is the specific internal energy, $P$ is the pressure‚Å† (the last two are determined using the equation of state; EOS), $P_{\rm mag} = b^{2}/2$ is the magnetic pressure, $b^{2} \equiv b^{\mu}b_{\mu}$, $b^{\mu} = \left(4\pi\right)^{-1/2}B^{\mu}_{(u)}$, and $B^{\mu}_{(u)}$ is the magnetic field measured by an observer comoving with the fluid:

$$
\begin{align}
B^{0}_{(u)} &= u_{i}B^{i}/\alpha,\\
B^{i}_{(u)} &= \frac{B^{i}/\alpha + B^{0}_{(u)}u^{i}}{u^{0}},
\end{align}
$$

with $B^{i}$ the magnetic field.

When considering advanced, tabulated EOSs, hydrodynamic quantities are typically given as functions of the baryonic density $\rhob\equiv \nb\mb$, where $\mb$ is the baryon mass, the temperature $T$, and the electron fraction

$$
\ye = \frac{\ne}{\nb} = \frac{\ne}{\rhob/\mb} = \frac{\ne\mb}{\rhob},
$$

where $\ne$ is the lepton number density. The temperature is typically not evolved explicitly in time, but instead determined from other hydrodynamical quantities using the EOS. The electron fraction, on the other hand, is then evolved using the conservation of lepton number density equation

$$
\nabla_{\mu}\left(\ne u^{\mu}\right) = 0.
$$

We note that evolution equations for the baryonic density and electron fraction are the ones typically used in numerical implementation. These are trivially obtained from the baryon and lepton number conservation equations by using the definitions of $\rhob$ and $\ye$ and are given by, respectively,

$$
\begin{align}
\nabla_{\mu}\left(\rhob u^{\mu}\right) &= 0,\\
\nabla_{\mu}\left(\rhob\ye u^{\mu}\right) &= 0.
\end{align}
$$

The GRMHD equations are thus

$$
\boxed{
\begin{split}
\nabla_{\mu}\left(\rhob u^{\mu}\right) &= 0\\
\nabla_{\mu}\left(\rhob\ye u^{\mu}\right) &= 0\\
\nabla_{\mu}T^{\mu\nu} &= 0\\
\nabla_{\nu}\Fdual^{\mu\nu} &= 0
\end{split}
}\, .
$$

## Step 1.b: An overview of the neutrino leakage scheme
$$\label{neutrino_leakage}$$

There are different approaches for including neutrino effects in the coalescence of binary neutron stars (BNS). For example, the observation that simulations typically model $\mathcal{O}(10)$ ms of the merging event can lead one to assume that transport effects by the difussion of neutrinos are of minor importance, with negligible neutrino momentum transfers. These are the core assumptions made by [Ruffert *et al.* (1996)](http://adsabs.harvard.edu/pdf/1996A%26A...311..532R) (see also [O'Connor & Ott (2010)](https://arxiv.org/abs/0912.2393), [Galleazi *et al.* (2013)](https://arxiv.org/abs/1306.4953), [Siegel & Metzger (2018)](https://arxiv.org/abs/1711.00868), and [Murguia-Berthier *et al.* (2021)](https://arxiv.org/abs/2106.05356)), where neutrino physics are included via a *neutrino leakage scheme*.

A neutrino leakage scheme takes into account possible equilibrations of neutrinos and the opaque, hot stellar matter to describe local neutrino sinks and sources. These, in turn, lead to variations in the number of leptons, which result in heating or cooling of the system. There is a variety of neutrino emission mechanisms that are important at different densities and temperatures within nuclear matter. The most important of them are (i) electron and positron capture on nucleons ([$\beta$-process](https://en.wikipedia.org/wiki/Beta_decay)); (ii) [electron-positron pair annihilation](https://en.wikipedia.org/wiki/Annihilation); (iii) transverse plasmon decay. We now provide an overview of these reactions, while details of how to compute the neutrino emission rates are postponed until [Step 1.b.i](#neutrino_rates) below.

The simplest and most powerful neutrino emission from hot ($T\gtrsim5\, {\rm MeV}$), dense ($\rhob\gtrsim10^{13}\, {\rm g\, cm^{-3}}$) nuclear matter comes from [Urca processes](https://en.wikipedia.org/wiki/Urca_process), in which electron neutrinos (antineutrinos) are produced via [electron (positron) capture](https://en.wikipedia.org/wiki/Electron_capture) by [protons](https://en.wikipedia.org/wiki/Proton) ([neutrons](https://en.wikipedia.org/wiki/Neutron)), i.e.,

$$
\begin{alignat}{3}
\ee &+ p &&\to n &&+ \nue,\\
\ae &+ n &&\to p &&+ \anue.
\end{alignat}
$$

These processes drive nucleons towards $\beta$-equilibrium, where the [chemical potential](https://en.wikipedia.org/wiki/Chemical_potential) of electron neutrinos and antineutrinos are zero, $\mu_{\nue}=0=\mu_{\anue}$, and the rate of both reactions is the same, leaving the matter composition unchanged.

At low densities ($\rhob\lesssim10^{10}\, {\rm g\, cm^{-3}}$) and high temperatures (nondegenerate nuclear matter), [annihilation of electron-positron pairs](https://en.wikipedia.org/wiki/Annihilation), i.e.,

$$
\ee + \ae \to \nui + \anui,
$$

becomes an efficient neutrino mechanism. Here, $\nui = \{\nue,\nux\}$ and $\anui = \{\anue,\anux\}$, where $x$ represents either the [muon](https://en.wikipedia.org/wiki/Muon) or the [tau](https://en.wikipedia.org/wiki/Tau_(particle)). In [Step 1.c](#neutrino_rates) we will discuss how to accurately calculate the neutrino emission rates from these processes using the temperature and electron fraction.

At intermediate densities ($10^{10}\, {\rm g\, cm^{-3}}\lesssim\rhob\lesssim10^{13}\, {\rm g\, cm^{-3}}$) and high temperatures, a major neutrino is [plasmon](https://en.wikipedia.org/wiki/Plasmon) decay. Recall that plasmons are quanta of electromagnetic field in a plasma that can be either [longitudinally](https://en.wikipedia.org/wiki/Longitudinal_wave) and [transversely](https://en.wikipedia.org/wiki/Transverse_wave) [polarized](https://en.wikipedia.org/wiki/Polarization_(waves). In the regime of
our simulations, the longitudinal polarization can be neglected and the plasmon decay written as

$$
\gamma \to \nui + \anui.
$$

where $\gamma$ represents the plasmon. 

At high density ($\rhob\gtrsim10^{13}\, {\rm g\, cm^{-3}}$) and very high temperatures ($T\gtrsim10\, {\rm MeV}$), neutrinos cannot freely escape but can only diffuse via a random walk over a time scale $t^{\rm diff}$ (see e.g., [Galleazi *et al.* (2013)](https://arxiv.org/abs/1306.4953)) given by (cf. Eq. 47 in [Murguia-Berthier *et al.* (2021)](https://arxiv.org/abs/2106.05356))

$$
t^{\rm diff} = D^{\rm diff} \kappa_{\nui}^{-1} \tau^{2},
$$

where $\kappa_{\nui}$ is the energy averaged opacity of $\nui$ and $\tau$ is the optical depth, defined as the integral of the inverse [mean free path](https://en.wikipedia.org/wiki/Mean_free_path) along a line of sight (see [Step 1.c](#neutrino_rates) below). In order to compute the mean free path, we take into account neutrino absorption and scattering processes with surrounding matter. We take into account the capture of electron neutrinos and antineutrinos by neutrons and protons, i.e.,

$$
\begin{align}
\nue + n &\to \ee + p,\\
\anue + p &\to \ae + n,\\
\end{align}
$$

and neutrino scattering off free nucleons, i.e.,

$$
\begin{align}
\nui + p &\to \nui + p,\\
\nui + n &\to \nui + n,\\
\anui + p &\to \anui + p,\\
\anui + n &\to \anui + n.
\end{align}
$$

The effective absorption/emission rate of neutrinos implies an effective change in lepton number and in heating/cooling of the system. These lead to modifications of our GRMHD equations in the form of source terms $\RR$ and $\QQ$, respectively, which enter the GRMHD equations as:

$$
\boxed{
\begin{split}
\nabla_{\mu}\left(\rhob u^{\mu}\right) &= 0\\
\nabla_{\mu}\left(\rhob\ye u^{\mu}\right) &= \color{red}{\RR}\\
\nabla_{\mu}T^{\mu\nu} &= \color{red}{\QQ u^{\nu}}\\
\nabla_{\nu}\Fdual^{\mu\nu} &= 0
\end{split}
}\, .
$$

The total absorption/emission rate for each neutrino species, $\RR_{\nui}$, is given by (cf. Eq. 43 of [Murguia-Berthier *et al.* (2021)](https://arxiv.org/abs/2106.05356))

$$
\RR_{\nui} \equiv \RR_{\nui}^{\beta} + \RR_{\nui}^{ee} + \RR_{\nui}^{\gamma},
$$

where $\RR_{\nui}^{\beta}$, $\RR_{\nui}^{ee}$, and $\RR_{\nui}^{\gamma}$ are the absorption/emission rates of the $\beta$-processes, electron-positron pair annihilation, and plasmon decay, respectively. Analogously, the corresponding energy loss rate, $\QQ_{\nui}$, is given by

$$
\QQ_{\nui} \equiv \QQ_{\nui}^{\beta} + \QQ_{\nui}^{ee} + \QQ_{\nui}^{\gamma}.
$$

The effective emission/absorption and energy loss rates are given by an interpolation between the diffusive optically thick regime and the transparent optically thin regime (see Appendix B of [Ruffert *et al.* (1996)](http://adsabs.harvard.edu/pdf/1996A%26A...311..532R) and Eqs. 22 of [Siegel & Metzger (2018)](https://arxiv.org/abs/1711.00868)),

$$
\begin{align}
\RR^{\rm eff}_{\nui} &= \RR_{\nui}\left(1 + \frac{t^{\rm diff}_{\nui}}{t^{{\rm em},\RR}_{\nui}}\right)^{-1},\\
\QQ^{\rm eff}_{\nui} &= \QQ_{\nui}\left(1 + \frac{t^{\rm diff}_{\nui}}{t^{{\rm em},\QQ}_{\nui}}\right)^{-1},
\end{align}
$$

where (cf. Eq. 23 of [Siegel & Metzger (2018)](https://arxiv.org/pdf/1711.00868.pdf))

$$
t^{\rm diff}_{\nui} = D^{\rm diff}\kappa_{\nui}^{-1}\tau_{\nui}^{2},
$$

denotes the local diffusion timescales, $\tau_{\nui}$ are the corresponding optical depths (see below), and $D^{\rm diff} = 6$ is a normalization factor. Furthermore,

$$
t^{{\rm em},\RR}_{\nui} = \frac{\RR_{\nui}}{n_{\nui}}\quad\text{and}\quad t^{{\rm em},\QQ}_{\nui} = \frac{\QQ_{\nui}}{e_{\nui}},
$$

are the local neutrino numbers and energy emission timescales, where $e_{\nui}$ refers to the neutrino energy densities.

Finally, once the opacities and optical depths are known, the source terms $\RR$ and $\QQ$ can be computed via (cf. Eqs. B24 and B25 of [Ruffert *et al.* (1996)](http://adsabs.harvard.edu/pdf/1996A%26A...311..532R))

$$
\boxed{
\begin{split}
\RR &\equiv -\RR^{\rm eff}_{\nue} + \RR^{\rm eff}_{\anue}\\
\QQ &\equiv -\QQ^{\rm eff}_{\nue} + \QQ^{\rm eff}_{\anue} + \QQ^{\rm eff}_{\nux}
\end{split}
}\ ,
$$

where the expressions for $\RR^{\rm eff}_{\anui}$ and $\QQ^{\rm eff}_{\anui}$ are identical to those of $\RR^{\rm eff}_{\nui}$ and $\QQ^{\rm eff}_{\nui}$ with $\nui\to\anui$.
