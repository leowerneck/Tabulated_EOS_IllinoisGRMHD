## Step 1.a: General relativistic magnetohydrodynamics \[Back to [Top](#toc)\]
$$\label{grmhd_equations}$$

The standard general relativistic [magnetohydrodynamics](https://en.wikipedia.org/wiki/Magnetohydrodynamics) (GRMHD) are given by (see e.g., [Duez *et al.* (2005)](https://arxiv.org/pdf/astro-ph/0503420.pdf) for an excellent overview)

$$
\newcommand{\RR}{\mathcal{R}}
\newcommand{\QQ}{\mathcal{Q}}
\newcommand{\nui}{\nu_{i}}
\newcommand{\anui}{\bar{\nu}_{i}}
\newcommand{\nux}{\nu_{\rm x}}
\newcommand{\anux}{\bar{\nu}_{\rm x}}
\newcommand{\nue}{\nu_{\rm e}}
\newcommand{\anue}{\bar{\nu}_{\rm e}}
\newcommand{\mb}{m_{\rm b}}
\newcommand{\me}{m_{\rm e}}
\newcommand{\nb}{n_{\rm b}}
\newcommand{\ne}{n_{\rm e}}
\newcommand{\rhob}{\rho_{\rm b}}
\newcommand{\ye}{Y_{\rm e}}
\newcommand{\Fdual}{{}^{*}\!F}
\newcommand{\ee}{e^{-}}
\newcommand{\ae}{e^{+}}
\begin{align}
\nabla_{\mu}\left(\nb u^{\mu}\right) &= 0,\\
\nabla_{\mu}T^{\mu\nu} &= 0,\\
\nabla_{\nu}\Fdual^{\mu\nu} &= 0.
\end{align}
$$

These are conservation of baryon number, energy and momentum conservation, and homogeneous [Maxwell's equations](https://en.wikipedia.org/wiki/Maxwell%27s_equations), respectively. In the above, $\nabla_{\mu}$ is the covariant derivative associated with the spacetime metric $g_{\mu\nu}$, $\nb$ is the baryon number density, $u^{\mu}$ is the fluid four-velocity, and $\Fdual^{\mu\nu} = \frac{1}{4}\epsilon^{\mu\nu\rho\sigma}F_{\rho\sigma}$ is the dual of the [Faraday tensor](https://en.wikipedia.org/wiki/Electromagnetic_tensor) $F^{\mu\nu}$ ($\epsilon^{\mu\nu\rho\sigma}$ is the [Levi-Civita tensor](https://en.wikipedia.org/wiki/Levi-Civita_symbol)). $T^{\mu\nu}$ is the energy-momentum tensor of a magnetized fluid and is given by (cf. Eq. 8 of [Etienne *et al.*(2015)](https://arxiv.org/pdf/1501.07276.pdf))

$$
T^{\mu\nu}_{\rm GRMHD} = \left(\rhob h + b^{2}\right)u^{\mu}u^{\nu} + \left(P + P_{\rm mag}\right)g^{\mu\nu} - b^{\mu}b^{\nu},
$$

where $h = 1 + \epsilon + P/\rhob$ is the specific enthalpy, $\epsilon$ is the specific internal energy, $P$ is the pressure⁠ (the last two are determined using the equation of state; EOS), $P_{\rm mag} = b^{2}/2$ is the magnetic pressure, $b^{2} \equiv b^{\mu}b_{\mu}$, $b^{\mu} = \left(4\pi\right)^{-1/2}B^{\mu}_{(u)}$, and $B^{\mu}_{(u)}$ is the magnetic field measured by an observer comoving with the fluid:

$$
\begin{align}
B^{0}_{(u)} &= u_{i}B^{i}/\alpha,\\
B^{i}_{(u)} &= \frac{B^{i}/\alpha + B^{0}_{(u)}u^{i}}{u^{0}},
\end{align}
$$

with $B^{i}$ the magnetic field.

When considering advanced, tabulated EOSs, hydrodynamic quantities are typically given as functions of the baryonic density $\rhob\equiv \nb\mb$, where $\mb$ is the baryon mass, the temperature $T$, and the electron fraction

$$
\ye = \frac{\ne}{\nb} = \frac{\ne}{\rhob/\mb} = \frac{\ne\mb}{\rhob},
$$

where $\ne$ is the lepton number density. The temperature is typically not evolved explicitly in time, but instead determined from other hydrodynamical quantities using the EOS. The electron fraction, on the other hand, is evolved using the conservation of lepton number density equation

$$
\nabla_{\mu}\left(\ne u^{\mu}\right) = 0.
$$

We note that typically one evolves the baryonic density and electron fraction instead of the baryon and lepton numbers, respectively. Evolution equations for $\rhob$ and $\ye$ are trivially obtained by using their definitions in the evolution equations for $\nb$ and $\ne$, respectively,

$$
\begin{align}
\nabla_{\mu}\left(\rhob u^{\mu}\right) &= 0,\\
\nabla_{\mu}\left(\rhob\ye u^{\mu}\right) &= 0.
\end{align}
$$

The GRMHD equations are thus

$$
\boxed{
\begin{split}
\nabla_{\mu}\left(\rhob u^{\mu}\right) &= 0\\
\nabla_{\mu}\left(\rhob\ye u^{\mu}\right) &= 0\\
\nabla_{\mu}T^{\mu\nu} &= 0\\
\nabla_{\nu}\Fdual^{\mu\nu} &= 0
\end{split}
}\, .
$$

<font color='red'>**Note:**</font> In [O'Connor & Ott (2010)](https://arxiv.org/abs/0912.2393) it is suggested to evolve the electron fraction using the conservation equation above to obtain $\ye(t+\Delta t)$ and then compute

$$
\Delta\ye = \min[0,\bar{\ye}(\rhob)-\ye(t+\Delta t)].
$$

using $\Delta\ye$ as the total variation of $\ye$ from one MoL (sub)step to the next. In the above, $\bar{\ye}(\rhob)$ a (tabulated) profile of the electron fraction as a function the the baryonic density which, according to simulations results obtained by [Liebendörfer (2005)](https://arxiv.org/pdf/astro-ph/0504072.pdf), does not vary considerably during the simulation (i.e., $\bar{\ye}(t,\rhob)$ is nearly constant in time). If a tabulated profile is not known, [Liebendörfer (2005)](https://arxiv.org/pdf/astro-ph/0504072.pdf) suggests using

$$
\bar{\ye}(\rhob) = \bar{\ye}[x(\rhob)] = \frac{1}{2}\left(Y_{2}+Y_{1}\right) + \frac{x}{2}\left(Y_{2}-Y_{1}\right) + Y_{\rm c}\left[1 - |x| + 4|x|\left(|x|-\frac{1}{2}\right)\left(|x|-1\right)\right],
$$

where

$$
x(\rhob) = \max\left[-1,\min\left(1,\frac{2\log\rhob-\log\rho_{2}-\log\rho_{1}}{\log\rho_{2}-\log\rho_{1}}\right)\right],
$$

This corresponds to a fitting based on a piecewise linear approximation of a time-dependent tabulated template, $\bar\ye(\rhob)$, with a piecewise cubic correction. They also provide two suggestions for the fitting parameters $\left(Y_{\rm c},Y_{1},Y_{2},\rho_{1},\rho_{2}\right)$, which are given according to Table I in [Liebendörfer (2005)](https://arxiv.org/abs/astro-ph/0504072), i.e.,

* Model N13: $Y_{\rm c}=0.035$, $Y_{1}=0.5$, $Y_{2}=0.285$, $\rho_{1}=2\times10^{7}\ {\rm g\,cm^{-3}}$, $\rho_{2}=2\times10^{13}\ {\rm g\,cm^{-3}}$;
* Model G15: $Y_{\rm c}=0.035$, $Y_{1}=0.5$, $Y_{2}=0.278$, $\rho_{1}=3\times10^{7}\ {\rm g\,cm^{-3}}$, $\rho_{2}=2\times10^{13}\ {\rm g\,cm^{-3}}$.

In [Liebendörfer *et al.* (2005)](https://arxiv.org/abs/astro-ph/0310662), the authors study stellar core collapse  and describe these models as being: a simplified model that includes the core neutrino physics, but assumes Newtonian gravity (N13) and a complete model that includes general relativity (G15).

<font color='red'>**We note here that this was not used in IllinoisGRMHD, but support for both the tabulated profile and the fitting formula are available within ZelmaniLeak.**</font>

## Step 1.b: An overview of the neutrino leakage scheme \[Back to [Top](#toc)\]
$$\label{neutrino_leakage}$$

There are different approaches for including [neutrino](https://en.wikipedia.org/wiki/Neutrino) effects in the [coalescence](https://en.wikipedia.org/wiki/Neutron_star_merger) of binary [neutron stars](https://en.wikipedia.org/wiki/Neutron_star). For example, the observation that simulations typically model $\mathcal{O}(10)$ ms of the merging event can lead one to assume that transport effects by the difussion of neutrinos are of minor importance, with negligible neutrino momentum transfers. These are the core assumptions made by [Ruffert *et al.* (1996)](http://adsabs.harvard.edu/pdf/1996A%26A...311..532R) (see also [O'Connor & Ott (2010)](https://arxiv.org/abs/0912.2393), [Galleazi *et al.* (2013)](https://arxiv.org/abs/1306.4953), [Siegel & Metzger (2018)](https://arxiv.org/abs/1711.00868), and [Murguia-Berthier *et al.* (2021)](https://arxiv.org/abs/2106.05356)), where neutrino physics are included via a *neutrino leakage scheme*.

A neutrino leakage scheme takes into account possible equilibrations of neutrinos and the opaque, hot stellar matter to describe local neutrino sinks and sources. These, in turn, lead to variations in the number of leptons, which result in heating or cooling of the system. There is a variety of neutrino emission mechanisms that are important at different densities and temperatures within nuclear matter. The most important of them are (i) electron and positron capture by nucleons ([$\beta$-process](https://en.wikipedia.org/wiki/Beta_decay)); (ii) [electron-positron pair annihilation](https://en.wikipedia.org/wiki/Annihilation); (iii) transverse plasmon decay. We now provide an overview of these reactions, postponing the details on how to compute the neutrino opacities and emission rates to [Step 1.b.i](#neutrino_opacities) and [Step 1.b.ii](#neutrino_rates), respectively.

The simplest and most powerful neutrino emission from hot ($T\gtrsim5\, {\rm MeV}$), dense ($\rhob\gtrsim10^{13}\, {\rm g\, cm^{-3}}$) nuclear matter comes from [Urca processes](https://en.wikipedia.org/wiki/Urca_process), in which electron neutrinos (antineutrinos) are produced via [electron (positron) capture](https://en.wikipedia.org/wiki/Electron_capture) by [protons](https://en.wikipedia.org/wiki/Proton) ([neutrons](https://en.wikipedia.org/wiki/Neutron)), i.e.,

$$
\begin{alignat}{3}
\ee &+ p &&\to n &&+ \nue,\\
\ae &+ n &&\to p &&+ \anue.
\end{alignat}
$$

These processes drive nucleons towards $\beta$-equilibrium, where the [chemical potential](https://en.wikipedia.org/wiki/Chemical_potential) of electron neutrinos and antineutrinos are zero, $\mu_{\nue}=0=\mu_{\anue}$, and the rate of both reactions is the same, leaving the matter composition unchanged.

At low densities ($\rhob\lesssim10^{10}\, {\rm g\, cm^{-3}}$) and high temperatures (nondegenerate nuclear matter), [annihilation of electron-positron pairs](https://en.wikipedia.org/wiki/Annihilation), i.e.,

$$
\ee + \ae \to \nui + \anui,
$$

becomes an efficient neutrino mechanism. Here, $\nui = \{\nue,\nux\}$ and $\anui = \{\anue,\anux\}$, where $x$ represents either the [muon](https://en.wikipedia.org/wiki/Muon) or the [tau](https://en.wikipedia.org/wiki/Tau_(particle)).

At intermediate densities ($10^{10}\, {\rm g\, cm^{-3}}\lesssim\rhob\lesssim10^{13}\, {\rm g\, cm^{-3}}$) and high temperatures, [plasmon](https://en.wikipedia.org/wiki/Plasmon) decay becomes major neutrino source. Recall that plasmons are quanta of electromagnetic field in a plasma that can be either [longitudinally](https://en.wikipedia.org/wiki/Longitudinal_wave) or [transversely](https://en.wikipedia.org/wiki/Transverse_wave) [polarized](https://en.wikipedia.org/wiki/Polarization_(waves). In the regime of
our simulations, the longitudinal polarization can be neglected and the plasmon decay written as

$$
\gamma \to \nui + \anui.
$$

where $\gamma$ represents the plasmon. 

At high density ($\rhob\gtrsim10^{13}\, {\rm g\, cm^{-3}}$) and very high temperatures ($T\gtrsim10\, {\rm MeV}$), neutrinos cannot freely escape but can only diffuse via a [random walk](https://en.wikipedia.org/wiki/Random_walk) over a time scale $t^{\rm diff}_{\nui}$ (see e.g., [Galleazi *et al.* (2013)](https://arxiv.org/abs/1306.4953)) given by (cf. Eq. 47 in [Murguia-Berthier *et al.* (2021)](https://arxiv.org/abs/2106.05356))

$$
t^{\rm diff}_{\nui} = D^{\rm diff} \kappa_{\nui}^{-1} \tau^{2},
$$

where $\kappa_{\nui}$ is the energy averaged opacity of $\nui$ and $\tau$ is the optical depth, defined as the integral of the inverse [mean free path](https://en.wikipedia.org/wiki/Mean_free_path) along the line of sight (see [Step 1.b.i](#neutrino_rates) below). In order to compute the mean free path, we consider neutrino absorption and scattering processes with surrounding matter. We take into account the capture of electron neutrinos and antineutrinos by neutrons and protons, respectively, i.e.,

$$
\begin{align}
\nue + n &\to \ee + p,\\
\anue + p &\to \ae + n,\\
\end{align}
$$

and neutrino scattering off free nucleons, i.e.,

$$
\begin{align}
\nui + p &\to \nui + p,\\
\nui + n &\to \nui + n,\\
\anui + p &\to \anui + p,\\
\anui + n &\to \anui + n.
\end{align}
$$

The effective absorption/emission rate of neutrinos implies an effective change in lepton number and in heating/cooling of the system. These lead to modifications of our GRMHD equations in the form of source terms $\RR$ and $\QQ$, respectively, which enter the GRMHD equations as:

$$
\boxed{
\begin{split}
\nabla_{\mu}\left(\rhob u^{\mu}\right) &= 0\\
\nabla_{\mu}\left(\rhob\ye u^{\mu}\right) &= \color{red}{\RR}\\
\nabla_{\mu}T^{\mu\nu} &= \color{red}{\QQ u^{\nu}}\\
\nabla_{\nu}\Fdual^{\mu\nu} &= 0
\end{split}
}\, .
$$

### Step 1.b.i: Neutrino opacities, optical depth, and diffusion time scales \[Back to [Top](#toc)\]
$$\label{neutrino_opacities}$$

We now provide the necessary formulas to compute neutrino opacities $\kappa_{\nui} = \lambda_{\nui}^{-1}$ (inverse mean free path), optical depths, and diffusion time scales.

The transport cross section for neutrino-nucleon scattering ($\nui + N \to \nui + N$) is given by (cf. Eq. A1 of [Ruffert *et al.* (1996)](http://adsabs.harvard.edu/pdf/1996A%26A...311..532R))

$$
\sigma_{s}(\nui,N) = C_{s,N}\sigma_{0}\left(\frac{\epsilon}{m_{e}c^{2}}\right)^{2},
$$

where $\sigma_{0} \approx 1.76 \times 10^{-44}\ {\rm cm^{2}}$, $\epsilon$ is the neutrino energy, $\me$ is the electron mass, $c$ is the speed of light, and

$$
C_{s,N} = \left\{
\begin{matrix}
(1+5\alpha^{2})/24, & \text{if }N=n,\\
[4(C_{V}-1)^{2}+5\alpha^{2}]/24, & \text{if }N=p.\\
\end{matrix}
\right.
$$

The degeneracy parameters for neutrinos ($\eta_{\nui}=\mu_{\nui}/T$) and antineutrinos ($\eta_{\anui}=\mu_{\anui}/T$) are set according to

$$
\begin{align}
\eta_{\nux} &= 0,\\
\eta_{\anux} &= 0,\\
\eta_{\nue} &= \eta_{\nue}^{\rm ceq}\left[1-\exp\left(-\tau_{\nue,\RR}\right)\right] + \eta_{\nue}^{0}\exp\left(-\tau_{\nue,\RR}\right),\\
\eta_{\anue} &= -\eta_{\nue}^{\rm ceq}\left[1-\exp\left(-\tau_{\anue,\RR}\right)\right] + \eta_{\anue}^{0}\exp\left(-\tau_{\anue,\RR}\right),
\end{align}
$$

where $\eta_{\nue}^{\rm ceq}$ is the degeneracy parameter for $\nue$ at chemical equilibrium with the stellar medium,

$$
\eta_{\nue}^{\rm ceq} = -\eta_{\anue}^{\rm ceq} = \eta_{\rm e} + \eta_{\rm p} - \eta_{\rm n} - Q/T,
$$

where $Q = 1.2935\ {\rm MeV}$ is the [rest-mass-energy](https://en.wikipedia.org/wiki/Mass_in_special_relativity) difference between a neutron and a proton. To avoid divergent behavior at low densities we interpolate for $\nue$ and $\anue$ between the chemical equilibrium values at high optical depths and fixed values $\eta_{\nue}^{0}$ and $\eta_{\anue}^{0}$ in the limit of transparent matter. We use $\eta_{\nue}^{0} = \eta_{\anue}^{0} = 0$, and the interpolation is expressed in terms of the optical depths $\tau_{\nue,\RR}$ and $\tau_{\anue,\RR}$ (see below). Note that computation of the optical depths requires knowledge of the neutrino opacities and thus require a step where $\eta_{\nue}$ and $\eta_{\anue}$ are computed from $\rhob$ (<font color=red>FIXME: how? In principle we should be able to estimate them using the table by computing $\mu_{\nui}$ from $(\rhob,\ye,T)$ and then the respective degeneracy parameters, but this is not explicitly mentioned in the paper</font>).

The scattering opacities are then computed using (cf. Eq. A6 of [Ruffert *et al.* (1996)](http://adsabs.harvard.edu/pdf/1996A%26A...311..532R))

$$
\kappa_{{\rm s},j}(\nui,N) = C_{s,N}\sigma_{0}N_{A}\rhob Y_{NN}\left(\frac{T}{\me c^{2}}\right)^{2}\frac{F_{4+j}(\eta_{\nui})}{F_{2+j}(\eta_{\nui})},
$$

where $j=0$ for neutrino-number transport ($\RR$) and $j=1$ for neutrino-energy transport ($\QQ$), $N_{A}$ is [Avogadro's number](https://en.wikipedia.org/wiki/Avogadro_constant), and

$$
Y_{NN} = \frac{Y_{N}}{1 + \frac{2}{3}\max(\eta_{N},0)},
$$

with $N$ representing the neutron or the proton and $Y_{\rm n} = 1 - \ye$ and $Y_{\rm p}=\ye$ the neutron and proton fractions, respectively.

The cross section for $\nue$ absorption onto a neutron ($\nue + n \to \ee + p$) is given by (cf. Eq. A9 of [Ruffert *et al.* (1996)](http://adsabs.harvard.edu/pdf/1996A%26A...311..532R))

$$
\sigma_{\rm a}(\nue,n) = \frac{1+3\alpha^{2}}{4}\sigma_{0}\left(\frac{\epsilon}{\me c^{2}}\right)^{2}\left[1-f_{\nue}\right]_{\rm ec},
$$

while the cross section for $\anue$ absorption onto a proton ($\anue + p \to \ae + n$) is given by (cf. Eq. A10 of [Ruffert *et al.* (1996)](http://adsabs.harvard.edu/pdf/1996A%26A...311..532R))

$$
\sigma_{\rm a}(\anue,p) = \frac{1+3\alpha^{2}}{4}\sigma_{0}\left(\frac{\epsilon}{\me c^{2}}\right)^{2}\left[1-f_{\anue}\right]_{\rm pc}.
$$

Expressions for $\left[1-f_{\nue}\right]_{\rm ec}$ and $\left[1-f_{\anue}\right]_{\rm pc}$ are given in [Step 1.b.ii](#neutrino_rates) below. Performing the spectral averaging we find the opacities for absorption processes are given by

$$
\kappa_{{\rm a},j}(\nue,n) = \frac{1+3\alpha^{2}}{4}\sigma_{0}N_{A}\rhob Y_{np} \left(\frac{T}{\me c^{2}}\right)^{2} \frac{F_{4+j}(\eta_{\nue})}{F_{2+j}(\eta_{\nue})}\left[1-f_{\nue}\right]_{\rm ec},
$$

and

$$
\kappa_{{\rm a},j}(\anue,p) = \frac{1+3\alpha^{2}}{4}\sigma_{0}N_{A}\rhob Y_{pn} \left(\frac{T}{\me c^{2}}\right)^{2} \frac{F_{4+j}(\eta_{\anue})}{F_{2+j}(\eta_{\anue})}\left[1-f_{\anue}\right]_{\rm pc},
$$

where

$$
Y_{np} = \frac{2\ye-1}{e^{-\hat\eta}-1},
$$

and

$$
Y_{pn} = e^{-\hat\eta}Y_{np}.
$$

We can now compute total transport opacities for $\nue$,

$$
\boxed{\kappa_{{\rm total},j}(\nue) = \kappa_{{\rm s},j}(\nue,n) + \kappa_{{\rm s},j}(\nue,p) + \kappa_{{\rm a},j}(\nue,n)}\ ,
$$

for $\anue$,

$$
\boxed{\kappa_{{\rm total},j}(\anue) = \kappa_{{\rm s},j}(\anue,n) + \kappa_{{\rm s},j}(\anue,p) + \kappa_{{\rm a},j}(\anue,p)}\ ,
$$

and for $\nux$

$$
\boxed{\kappa_{{\rm total},j}(\nux) = \kappa_{{\rm s},j}(\nux,n) + \kappa_{{\rm s},j}(\nux,p)}\ .
$$

The optical depths along paths $[s_{1},s_{2}]$ can then be computed using

$$
\boxed{\tau_{\nui,j}([s_{1},s_{2}]) = \int_{s_{1}}^{s_{2}}ds \kappa_{{\rm total},j}(\nui)}\ ,
$$

then yielding the diffusion time scale

$$
\boxed{t^{\rm diff}_{\nui,j}([s_{1},s_{2}]) \approx \frac{3(s_{2}-s_{1})}{c}\tau_{\nui,j}([s_{1},s_{2}])}\ .
$$

### Step 1.b.ii: Neutrino emission/absorption and heating/cooling rates \[Back to [Top](#toc)\]
$$\label{neutrino_rates}$$

We start out discussion by looking at the electron capture:

$$
\ee + p \to n + \nue.
$$

The emission and cooling rates of this process, $\RR_{\rm ec}(\nue)$ and $\QQ_{\rm ec}(\nue)$, are given by (cf. Eqs. A5 and A4 of [Galleazi *et al.* (2013)](https://arxiv.org/abs/1306.4953))

$$
\begin{align}
\RR_{\rm ec}(\nue) &= \nb^{-1}\beta\,\eta_{\rm np}T^{5}F_{4}(\eta_{\rm e})\left[1-f_{\nue}\right]_{\rm ec},\\
\QQ_{\rm ec}(\nue) &= \nb^{-1}\beta\,\eta_{\rm np}T^{6}F_{5}(\eta_{\rm e})\left[1-f_{\nue}\right]_{\rm ec}.
\end{align}
$$

Similarly, for the positron capture:

$$
\ae + n \to p + \anue,
$$

we have (cf. Eqs. A3 and A2 of [Galleazi *et al.* (2013)](https://arxiv.org/abs/1306.4953))

$$
\begin{align}
\RR_{\rm pc}(\anue) &= \nb^{-1}\beta\,\eta_{\rm np}T^{5}F_{4}(-\eta_{\rm e})\left[1-f_{\anue}\right]_{\rm pc},\\
\QQ_{\rm pc}(\anue) &= \nb^{-1}\beta\,\eta_{\rm np}T^{6}F_{5}(-\eta_{\rm e})\left[1-f_{\anue}\right]_{\rm pc}.
\end{align}
$$

In the above, $\left[1-f_{\nui}\right]$ is the blocking factor for a neutrino [Fermi-Dirac distribution](https://en.wikipedia.org/wiki/Fermi–Dirac_statistics) $f_{\nui}$ and $\beta$ is a constant defined as

$$
\beta = \frac{\pi}{h^{3}c^{2}}\frac{1+3\alpha^{2}}{\me^{2} c^{4}}\sigma_{0},\\
$$

where $h$ is the [Planck constant](https://en.wikipedia.org/wiki/Planck_constant), $c$ is the [speed of light](https://en.wikipedia.org/wiki/Speed_of_light), $\me$ is the electron mass, $\sigma_{0} \approx 1.705 \times 10^{-44}\ {\rm cm^{2}}$ ([Ruffert *et al.* (1996)](http://adsabs.harvard.edu/pdf/1996A%26A...311..532R) use $\sigma_{0} \approx 1.76 \times 10^{-44}\ {\rm cm^{2}}$), the weak axial-vector coupling constant is $\alpha\approx1.23$ ([Ruffert *et al.* (1996)](http://adsabs.harvard.edu/pdf/1996A%26A...311..532R) use $\alpha\approx1.25$), $\eta_{i}\equiv\mu_{i}/T$ are the degeneracy parameters for different particle species, and

$$
n_{\rm np} = \frac{n_{\rm n}-n_{\rm p}}{e^{\hat\eta}-1},
$$

is the Pauli blocking factor, where $n_{\rm n}$ and $n_{\rm p}$ are the neutron and proton number densities, respectively, and $\hat\eta=\eta_{\rm n}-\eta_{\rm p}$ is the difference between neutron and proton relativistic degeneracy parameters. For densities smaller than $2\times10^{12}\ {\rm g\,cm^{-3}}$, we use $n_{\rm np}\approx n_{\rm n}$ instead. Temperatures are given in MeV.

$F_{N}$ are relativistic [Fermi-Dirac integrals](https://en.wikipedia.org/wiki/Complete_Fermi–Dirac_integral) (without the factor of $\Gamma^{-1}$) of order $N$,

$$
F_{N}(\eta) = \int_{0}^{\infty}\frac{x^{N}dx}{e^{x-\eta}+1}.
$$

These integrals cannot be computed analytically, and although they could easily be evaluated numerically, it is suggested to use the approximate expressions provided by [Takahashi, Eid, and Willebrandt (1978)](http://adsabs.harvard.edu/pdf/1978A%26A....67..185T) to evaluate the integrals, i.e.,

$$
F_{N}(\eta) = \left\{
\begin{matrix}
\frac{\eta^{N+1}/(N+1) + \sum_{r=1}^{(N+1)/2} 2\cdot {}_{N}C_{2r-1}\eta^{N+1-2r}F_{2r-1}(0)}{1 - (-1)^{N}\exp\left[-2N\eta F_{N-1}(0)/F_{N}(0)\right]},\ \text{if }\eta>10^{-3},\\
N!e^{\eta}\left[1+\left(\frac{N!}{F_{N}(0)}-1\right)\exp(A)\right]^{-1},\ \text{if }\eta<10^{-3},\\
\end{matrix}
\right.
$$

where

$$
{}_{N}C_{2r-1} = \frac{N!}{(2r-1)!(K-2r+1)!},
$$

and

$$
A = \left[\frac{1 - N F_{N-1}(0)/F_{N}(0)}{1-F_{N}(0)/N!}\right]\eta.
$$

The result for $N=0$ is known analytically,

$$
F_{0}(\eta) = \eta + \ln\left(1+e^{-\eta}\right).
$$

We also use the approximate expressions (cf. Eqs. A9 and A10 of [Galleazi *et al.* (2013)](https://arxiv.org/abs/1306.4953))

$$
\begin{align}
\left[1-f_{\nue}\right]_{\rm ec} &\approx \left\{1 + \exp\left[-\frac{F_{5}(\eta_{\rm e})}{F_{4}(\eta_{\rm e})}-\eta_{\nue}\right]\right\}^{-1},\\
\left[1-f_{\anue}\right]_{\rm pc} &\approx \left\{1 + \exp\left[-\frac{F_{5}(-\eta_{\rm e})}{F_{4}(-\eta_{\rm e})}-\eta_{\anue}\right]\right\}^{-1}.
\end{align}
$$

Next we look at the electron-positron pair annihilation leading to the production of an electron neutrino-antineutrino pair,

$$
\ae + \ee \to \nue + \anue.
$$

The neutrino emission rate for this process is given by (cf. Eq. A13 of [Galleazi *et al.* (2013)](https://arxiv.org/abs/1306.4953))

$$
\RR_{\ae\ee}(\nue,\anue) = \frac{(C_{1}+C_{2})_{\nue\anue}}{36\nb}\frac{\sigma_{0}c}{\me^{2}c^{4}}
\epsilon_{4}(\ae)\epsilon_{4}(\ee)\left[1-f_{\nue}(\eta_{\rm e},\eta_{\nue})\right]_{\ae\ee}\left[1-f_{\anue}(\eta_{\rm e},\eta_{\anue})\right]_{\ae\ee},
$$

where

$$
(C_{1}+C_{2})_{\nue\anue} = \left(C_{V}-C_{A}\right)^{2} + \left(C_{V}+C_{A}\right)^{2},
$$

with $C_{V}=0.962$ and $C_{A} = 0.5$. $\epsilon_{N}$ are energy moments of the Fermi-Dirac distribution for massive particles, given by (cf. Eq. A15 of [Galleazi *et al.* (2013)](https://arxiv.org/abs/1306.4953))

$$
\epsilon_{N}\bigl(e^{\mp}\bigr) = \frac{8\pi}{(hc)^{3}}T^{N}F_{N-1}\bigl(\pm\eta_{\rm e}\bigr).
$$

For the process involving heavy neutrinos, i.e.,

$$
\ae + \ee \to \nux + \anux,
$$

the emission rate is given by

$$
\RR_{\ae\ee}(\nux,\anux) = \frac{(C_{1}+C_{2})_{\nux\anux}}{9\nb}\frac{\sigma_{0}c}{\me^{2}c^{4}}
\epsilon_{4}(\ae)\epsilon_{4}(\ee)\left[1-f_{\anux}(\eta_{\rm e},\eta_{\nux})\right]_{\ae\ee},
$$

where $(C_{1}+C_{2})_{\nux\anux} = (C_{V}-C_{A})^{2} + (C_{V}+C_{A}-2)^{2}$ and (cf. Eq. A17 of [Galleazi *et al.* (2013)](https://arxiv.org/abs/1306.4953))

$$
\left[1-f_{\anux}(\eta_{\rm e},\eta_{\nux})\right]_{\ae\ee} \approx \left\{1+\exp\left[-\frac{1}{2}\left(\frac{F_{3}(\eta_{\rm e})}{F_{4}(\eta_{\rm e})} + \frac{F_{3}(-\eta_{\rm e})}{F_{4}(-\eta_{\rm e})}\right)-2\eta_{x}\right]\right\}^{-1}.
$$

The heating/cooling rate is given by (cf. Eq. A18 of [Galleazi *et al.* (2013)](https://arxiv.org/abs/1306.4953))

$$
\QQ_{\ae\ee}(\nui,\anui) = \RR_{\ae\ee}(\nui,\anui)\frac{\epsilon_{5}(\ee)\epsilon_{4}(\ae) + \epsilon_{4}(\ee)\epsilon_{5}(\ae)}{\epsilon_{4}(\ee)\epsilon_{4}(\ae)}.
$$

Lastly, we consider plasmon decay

$$
\begin{align}
\gamma &\to \nue + \anue,\\
\gamma &\to \nux + \anux,
\end{align}
$$

for which the emission rates are given by (cf. Eqs. A21 and A22 of [Galleazi *et al.* (2013)](https://arxiv.org/abs/1306.4953))

$$
\begin{align}
\RR_{\gamma}(\nue,\anue) &= \nb^{-1}\frac{\pi^{3}}{3\alpha_{\rm f}}C_{V}^{2}\frac{\sigma_{0}c}{\me^{2}c^{4}}\gamma^{6}\frac{T^{8}}{(hc)^{6}}\exp(-\gamma)(1+\gamma)\left[1+f_{\nue}(\eta_{\rm e},\eta_{\nue})\right]_{\gamma}\left[1+f_{\anue}(-\eta_{\rm e},\eta_{\anue})\right]_{\gamma},\\
\RR_{\gamma}(\nux,\anux) &= \nb^{-1}\frac{4\pi^{3}}{3\alpha_{\rm f}}(C_{V}-1)^{2}\frac{\sigma_{0}c}{\me^{2}c^{4}}\gamma^{6}\frac{T^{8}}{(hc)^{6}}\exp(-\gamma)(1+\gamma)\left[1+f_{\nux}(\eta_{\rm x},\eta_{\nux})\right]_{\gamma},
\end{align}
$$

where $\alpha_{\rm f}=1/137.036$ is the fine-structure constant and

$$
\gamma \approx f_{\rm p}\sqrt{\frac{1}{3}\left(\pi^{2}+3\eta_{\rm e}^{2}\right)},
$$

with $f_{\rm p} = 5.565\times10^{-2}$ the dimensionless plasma frequency. The blocking factor is given by (cf. Eq. A24 of [Galleazi *et al.* (2013)](https://arxiv.org/abs/1306.4953))

$$
\left[1-f_{\nui}(\eta_{\rm e},\eta_{\nui})\right]_{\gamma} \approx \left\{1 + \exp\left[-\left(1 + \frac{\gamma^{2}}{2(1+\gamma)}-\eta_{\nui}\right)\right]\right\}^{-1}.
$$

The heating/cooling rate is given by (cf. Eq. A25 of [Galleazi *et al.* (2013)](https://arxiv.org/abs/1306.4953))

$$
\QQ_{\gamma}(\nui,\anui) = \RR(\nui,\anui)T\left[1 + \frac{\gamma^{2}}{2(1+\gamma)}\right].
$$

Making use of the neutrino number densities

$$
n_{\nui} \equiv g_{\nui} \frac{4\pi}{(hc)^{3}}T^{3}F_{2}(\eta_{\nui}),
$$

and energy densities

$$
\epsilon_{\nui} \equiv g_{\nui}\frac{4\pi}{(hc)^{3}}T^{4}F_{3}(\eta_{\nui}),
$$

where $g_{\nue} = g_{\anue} = 1$ and $g_{\nux} = 4$, we can define the neutrino-number emission timescales as

$$
\left(t^{\rm loss}_{\nui,\RR}\right)^{-1} \equiv \frac{\RR_{\rm ec}(\nui) + \RR_{\ee\ae}(\nui) + \RR_{\gamma}(\nui)}{n_{\nui}} \equiv \frac{\RR_{\rm total}(\nui)}{n_{\nui}},
$$

and the neutrino-energy emission timescales as

$$
\left(t^{\rm loss}_{\nui,\QQ}\right)^{-1} \equiv \frac{\QQ_{\rm ec}(\nui) + \QQ_{\ee\ae}(\nui) + \QQ_{\gamma}(\nui)}{\epsilon_{\nui}} \equiv \frac{\QQ_{\rm total}(\nui)}{\epsilon_{\nui}},
$$

where, of course, the $\beta$-process contributions only apply to $\nue$ and $\anue$. Effective emission rates of neutrino number and energy can now be defined as functions of the relative size of the shortest diffusion timescale $t^{\rm diff}_{\nui,j}$ of a grid cell and the neutrino emission timescale $t^{\rm loss}_{\nui,j}$ of the cell:

$$
\begin{align}
\RR^{\rm eff}(\nui) &\equiv \frac{\RR_{\rm total}(\nui)}{1 + t^{\rm diff}_{\nui,\RR} \left(t^{\rm loss}_{\nui,\RR}\right)^{-1}},\\
\QQ^{\rm eff}(\nui) &\equiv \frac{\QQ_{\rm total}(\nui)}{1 + t^{\rm diff}_{\nui,\QQ} \left(t^{\rm loss}_{\nui,\QQ}\right)^{-1}}.
\end{align}
$$

In the (nearly) transparent, low-density regime, the optical depth to neutrino vanishes and the diffusion timescale is short compared to the timescale of direct neutrino loss. Therefore $\RR^{\rm eff}(\nui)\to\RR_{\rm total}(\nui)$ and $\QQ^{\rm eff}(\nui)\to\QQ_{\rm total}(\nui)$. On the other hand, in the opaque region the diffusion timescale becomes very long (and due to the increase of rates with density and temperature the direct loss timescale becomes extremely short, too). In this case, $\RR^{\rm eff}(\nui)\to n_{\nui}/t^{\rm diff}_{\nui,\RR}$ and $\QQ^{\rm eff}(\nui)\to \epsilon_{\nui}/t^{\rm diff}_{\nui,\QQ}$, i.e., the loss of neutrinos is then determined by the much slower diffusion process that depletes the equilibrium neutrino distributions only for a long time.

Finally, we define the source terms $\RR$ and $\QQ$ via

$$
\boxed{ \RR \equiv -\RR^{\rm eff}(\nue) + \RR^{\rm eff}(\anue) }\ ,
$$

and

$$
\boxed{ \QQ \equiv -\QQ^{\rm eff}(\nue) + \QQ^{\rm eff}(\anue) + \QQ^{\rm eff}(\nux) }\ .
$$

Additionally,

$$
\langle\epsilon\rangle_{\nui} = \frac{\QQ^{\rm eff}(\nui)}{\RR^{\rm eff}(\nui)},
$$

is the average energy of neutrinos of species $\nui$ emitted from a grid cell, and the average energy emitted by the system is the ratio of $\QQ^{\rm eff}(\nui)$ and $\RR^{\rm eff}(\nui)$ summed individually over the whole grid.

### Step 1.b.iii: Neutrino pressure \[Back to [Top](#toc)\]
$$\label{neutrino_pressure}$$

The `ZelmaniLeak` thorn also allows us to consider additional contributions to the total pressure and energy due to neutrinos. The neutrino pressure is given by (cf. Eq. (36) of [O'Connor & Ott (2010)](https://arxiv.org/abs/0912.2393))

$$
P_{\rm neutrinos} = \frac{4\pi}{3(hc)^{2}}T^{4}\bigl[F_{3}(\eta_{\nu}) + F_{3}(-\eta_{\nu})\bigr],
$$

and the neutrino energy by

$$
\epsilon_{\rm neutrinos} = \frac{3P_{\rm neutrinos}}{\rhob}.
$$

When considering these, the energy-momentum tensor is modified as (compare with Eq. (38) of [O'Connor & Ott (2010)](https://arxiv.org/abs/0912.2393))

$$
T^{\mu\nu} = T^{\mu\nu}_{\rm GRMHD} + \bigl(\rhob\epsilon_{\rm neutrinos} + P_{\rm neutrinos}\bigr)u^{\mu}u^{\nu} + P_{\rm neutrinos}g^{\mu\nu}.
$$

Furthermore, this radiation stress contributes source terms to the GRMHD equations so that (compare with Eqs. (37) of [O'Connor & Ott (2010)](https://arxiv.org/abs/0912.2393))

$$
\begin{align}
\big[\partial_{t}\tau\bigr]_{\rm neutrino\ source} &= -\sqrt{-g}v^{i}\partial_{i}P_{\rm neutrinos},\\
\big[\partial_{t}\tilde{S}_{i}\bigr]_{\rm neutrino\ source} &= -\sqrt{-g}\partial_{i}P_{\rm neutrinos},
\end{align}
$$

which are only applied to the right-hand sides of the GRMHD evolution equations when $\rhob > \rho_{\rm threshold}$, with $\rho_{\rm threshold}$ a freely specifiable parameter. In order to have a smooth transition between the regime where the neutrino pressure contributes to the evolution and the regime where it does not, in the numerical implementation the source term are multiplied by the transition function

$$
f(\rhob) = \frac{1}{2}\left[\tanh\left(\frac{\rhob-\rho_{\rm threshold}}{\rho_{\rm threshold}/10}\right)+1\right].
$$
